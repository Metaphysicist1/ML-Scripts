```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
num_set_1 = np.log2(np.arange(0,19,2))
num_set_2 = np.log(np.arange(0,19,2))

plt.xlabel("Log2 Values")
plt.ylabel("Log Values")
plt.title("Math Functions")

plt.plot(num_set_1,label='Log2',color='b',marker='^',linewidth=0.5)
plt.plot(num_set_2,label='Log',color='r',marker='*',linewidth=0.5)
plt.legend()
plt.savefig('log.png',dpi=500)
plt.show()
```
